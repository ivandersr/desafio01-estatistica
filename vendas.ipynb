```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mp
from datetime import datetime

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame(dict_faturamento)
```

```python
df_faturamento['valor'].mean()
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref']).dt.strftime('%m/%Y')
```

```python
# Gráfico de barras de faturamento por data
plt.figure(figsize=(14,6))
plt.bar(df_faturamento['data_ref'], df_faturamento['valor'])
ax = plt.gca()
ax.set_ylim([0,1100000])
ax.set_xlabel('Mês de referência')
ax.set_ylabel('Faturamento')
ax.set_title('Faturamento por Mês de referência')
ax.set_yscale('linear')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
# Gráfico de linhas de faturamento por data
plt.figure(figsize=(14,6))
plt.plot(df_faturamento['data_ref'], df_faturamento['valor'])
ax = plt.gca()
ax.set_ylim([0,1100000])
ax.set_xlabel('Mês de referência')
ax.set_ylabel('Faturamento')
ax.set_title('Faturamento por Mês de referência')
ax.set_yscale('linear')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
